```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_theme(style="darkgrid", palette = "pastel")
```

```python
df = pd.read_csv("datas/Sleep_Efficiency.csv", index_col = "ID", parse_dates=["Wakeup time", "Bedtime"])
df.info()
```

```python
df["Gender"].unique()
```

```python
sns.displot(data = df, x= "Sleep efficiency", hue = "Gender")
```

```python
df["Caffeine consumption"].value_counts()
```

```python
sa = pd.pivot_table(df,"Sleep efficiency", index = "Smoking status", columns = "Alcohol consumption")
sns.heatmap(sa, cmap = "crest")
```

```python
sns.histplot(data = df, y= "Sleep duration", x = "Sleep efficiency", hue = "Sleep duration", legend = False) #kind = 'kde' )
```

```python
df["Wakeup time"]=df["Wakeup time"].dt.time
df["Bedtime"]=df["Bedtime"].dt.time
```

```python
df["Wakeup time"].sort_values()
```

```python
df["Alcohol consumption"] = df["Alcohol consumption"].fillna(df["Alcohol consumption"].mean())

df["Alcohol consumption"] = df["Alcohol consumption"].astype('i').astype('object')
```
