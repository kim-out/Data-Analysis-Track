```python
#피보나치 수열 top down
def fib(n):
    if n==0 or n==1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
#메모이제이션(Memoization): 중복 계산 방지
memo = {0 : 1, 1 : 1}
def fib(n):
    if n in memo:
        return memo[n]
    else:
        result = fib(n-1) + fib(n-2)
        memo[n] = result
        return result
```

```python
#인접하지 않은 수들의 합의 최대값
data = [3, 4, 5, 6, 1, 2, 10]
result = []
def solution(data):
    if len(data) == 1:
        return data[0]
    result = [data[0], max(data[0], data[1])]
    for i in range(2, len(data)):
        result.append(max(result[i-1], result[i-2] + data[i]))
    return result[-1]
```

```python
n = 15
x = 1
cnt_list = []
cnt = 0

while x < n:
    x *= 3
    cnt += 1
if x == n:
    return cnt
else:
    x /= 3
    cnt -= 1
    while x < n:
        x *= 2
        cnt += 1
    if x == n:
        return cnt
    else: 
        cnt_list.append(cnt + n-x)
    
```

```python
n = int(input())
result = []
memo = []

def sol(n, cnt):      
    if n == 1:
        result.append(cnt)
        return
    if n%3 == 0 and n/3 not in memo:
        sol(n/3, cnt+1)
        memo.append(n/3)
    if n%2 == 0 and n/2 not in memo:
        sol(n/2, cnt+1)
        memo.append(n/2)
    sol(n-1, cnt+1)
    memo.append(n/2)

        
sol(n,0)
min(result)
```
