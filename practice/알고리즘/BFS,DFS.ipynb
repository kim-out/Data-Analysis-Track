```python
# 내 풀이(틀림)
n,m = map(int, input().split())
maps = []
for i in range(n):
    maps.append(list(map(int, input())))

def solution(n,m,maps):
    stack = [[0,0]]
    des = [n-1,m-1]
    moves = []
    move = 0
    while len(stack) > 0:
        now = stack.pop()
        cnt = len(stack)
        if now == des:
            moves.append(move+1)
            continue
        x = now[0]
        y = now[1]
        if x+1 <= n-1:
            if maps[x+1][y] == 1:
                stack.append([x+1, y])
                maps[x+1][y] = 2
        if x-1 >= 0:
            if maps[x-1][y] == 1:
                stack.append([x-1, y])
                maps[x-1][y] = 2
        if y+1 <= m-1:
            if maps[x][y+1] == 1:
                stack.append([x, y+1])
                maps[x][y+1] = 2
        if y-1 >= 0:
            if maps[x][y-1] == 1:
                stack.append([x, y-1])
                maps[x][y-1] = 2  
        if len(stack) == cnt:
            continue
        move += 1
    return min(moves)
solution(n,m,maps)       
 
```
