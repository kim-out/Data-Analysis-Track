```python
def split_multicolumn(series):
    l=[]
    series=series.fillna('')
    for i in series.unique():
        if ';' in i:
            l+=i.split(';')
    col = list(set(l))
    data = []
    for items in series.items():
        for i in col:
            if i in col:
                data.append(True)
            else:
                data.append(False)
    df= pd.DataFrame(data = np.ndarray(data).reshape(len(col),len(series)),columns = col, index = series.index)
    return df
```
