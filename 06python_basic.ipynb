```python
# key값에 올 수 있는 것: str, int, float, bool, tuple -> hashable
# key값에 올 수 없는 것: list, dictionary, 집합
# value값에는 모든 자료형 사용 가능
dic_var = {"key":"value",
           1:"value2",
           1.1:"value3",
           True:"value4",
           ("key",1,1.1,True):"value2"}
print(list(map(type,dic_var.keys())))
print(list(map(type,dic_var.values())))
```

```python
# 기존 리스트 전체 범위의 슬라이싱 객체는 다른 객체
test = [i for i in range(10)]
test2 = tuple(test)
print(id(test),id(test[:]))
```

```python
prac_result6 = [prac_result4[:] for _ in range(4)]
print("prac_result4의 id:", id(prac_result4))
for i, v in enumerate(prac_result6):
    print("prac_result6[",i,"] : ",id(v),sep="")
```

```python
old_list = list(range(10))
new_list = old_list[:]

print(id(old_list))
print(id(new_list))
print(old_list is new_list) # 객체 위치 비교(id함수의 결과, 주소값)
print(old_list == new_list) # 값을 비교
```

```python
# <참고> tuple, str은 전체범위 슬라이싱 결과의 주소값이 변하지 않을 수 있음
# immutable이기 때문에 새로 만들 필요가 없기 때문
old_tuple = tuple(range(10))
new_tuple = old_tuple[:]
print(old_tuple is new_tuple)
print(old_tuple == new_tuple)
```

```python
# 리스트 컴프리헨션2
# list()를 이용한 방식은 generator객체를 이용하므로 대괄호 방식에 비해 느림
list_d = list(i**3 for i in range(10))
print(list_d)
```

```python
# 리스트를 다른 변수에 할당하여 복사한 경우 주소값이 같음
# 원본의 값을 변경할 경우 대입한 리스트의 값도 변경
list_e = [i for i in range(10)]
print(list_e)
```

```python
# 위와 달리 copy()메서드를 이용해 리스트를 복제할 경우 새롭게 만든 객체이므로 주소가 다름
# 원본의 값을 변경해도 새로운 리스트 값에 영향을 미치지 않음
list_g = list(range(10))
list_h = list_g.copy()
print(id(list_g))
print(id(list_h))
print(list_g is list_h)
```

```python
list_g[0] = 999
print(list_g)
print(list_h)
```

```python
# 얕은 복사 (Shallow copy)
# 2차원 이상의 리스트는 내부 리스트가 메모리 주소를 담고 있는 참조 변수임
# 따라서 객체 주소만 복사해 오기 때문에, 내부 리스트의 메모리 주소는 그대로 유지
list_i = [[0]] * 3
print(list_i)
list_i[0][0] = 999
```

```python
print(list_i)
list_i[0][0] is list_i[1][0]
```

```python
# 깊은 복사 (Deep copy)
import copy
list_n = [[9,10],[11,12]]
list_o = copy.deepcopy(list_n)
print(list_n[0] is list_o)
```

```python
list_a=[[[1]],[[1]],[[1]]]
list_b=copy.deepcopy(list_a)
list_c=list_a.copy()
list_a[0][0][0]=999

print(list_b[0][0][0], list_c[0][0][0])
```
