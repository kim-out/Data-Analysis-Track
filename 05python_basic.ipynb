range함수 특징
  -> 리스트는 요소를 모두 저장해 두는 반면 range는 메모해두는 것과 비슷
for문은 iterable(시퀀스 자료형)의 elements를 하나씩 꺼냄
  ->반복 횟수가 정해져 있을 때, 반복 지시를 하거나 iterable 객체를 루핑(하나씩 꺼냄)
  참고>파이썬은 동적 타입, 자바는 정적 타입이기 때문에 자바는 for문 사용 시
       for(int i = 0; i<10; i++}()의 형태로 i의 타입을 명시해야 함


```python
for i in range(100):
    print(f"Hello world {i+1}")
    
```

```python
n=int(input("인사 몇 번 해드릴까요?"))
for i in range(n):
    print("Hello world")
```

```python
n=int(input("자연수 하나 입력해주세요:"))
for i in range(n,0,-1):
    r=(str(i)+"\t")*i
    print(r)
```

```python
for i in range(10):
    if i==0 or i==9:
        for j in range(10):
            print("*",end=' ')
        print()
    elif i==5:
        print("*     가현ㅎㅇ     *")
    else:
        for j in range(10):
            if j==0 or j==9:
                print("*",end=' ')
            else:
                print("  ",end='')
        print()
```

```python
n=int(input("1부터 입력한 자연수까지의 모든 홀수의 합을 구합니다"))
r=0
for i in range(1,n,2):
    r+=i
    if i==n or i==n-1:
        print(f"{i} = ",end='')
    else:
        print(f"{i} + ",end='')
print(r)
```

```python
n=int(input("계승을 구할 숫자를 입력하세요."))
r=1
for _ in range(1,n+1):
    r*=_
print(r)
```

```python
for _ in ["시","미","라","레","솔","도","파"]:
    print(_)
```

```python
for _ in "일이삼사오육칠팔구십":
    print(_, end=" ")
```

```python
# reversed(시퀀스) -> 뒤집은 값을 반환만 하고 해당 시퀀스를 변경하진 않음
#      =변수[::-1]
# reverse() -> 함수는 요소를 뒤집어 주며 반환값은 없음
for _ in reversed("일이삼사오육칠팔구십"):
    print(_, end=" ")
```

```python
x=[10,-42,25,103,21,77,24]
for _ in x:
    print(_*10,end=' ')
```
