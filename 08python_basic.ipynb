parameter = 함수의 매개 변수
argument = 함수의 변수에 넣는 값

함수(위치 전용, /, *, 키워드 전용, *가변 위치, ** 가변 키워드)
가변 위치 -> dict 함수 처럼 변수의 개수를 정해둘 수 없는 경우

```python
# 함수의 parameter가 default값을 갖지 않는 경우, 함수 호출 시 argument전달 필수
# 위치 인수는 순서에 영향을 받지만 키워드 인수는 상관X
# greeting(name="파이썬",32) -> 에러, "파이썬"은 키워드 인수로, 위치 인수의 뒤에 와야 함
```

```python
# 위치 전용 인자
def posonly(posonly, /): # '/'의 좌측
    print(posonly)
posonly("값만 입력해야 합니다.")
```

```python
# 키워드 전용 인자
def keyonly(*, keyonly = "default): # '*'의 우측
    print(keyonly)
keyonly()
```

```python
keyonly(keyonly="키워드로만 입력해야 합니다.")
```

```python
# 가변 위치 인자 -> 위치 인자를 개수에 상관없이 전달받음
def var_positional(*args):  # *변수
    print(type(args))
    return sum([_ for _ in args])
print(var_positional(1,2,3,4,5))
var_positional()
```

```python
# 가변 키워드 인자 -> 키워드 인수를 개수에 상관없이 전달 받음
def var_keyword(**kargs):
    print(type(kargs))
    return kargs
var_keyword({})
```

```python
print(var_keyword(key="value", key2="value2"))
# 키워드 인수들을 dict로 packing하여 인자에 전달
```

```python
def odd_even(number):
    """number가 짝수면 "짝수", 홀수면 "홀수" 반환"""
    if type(number)!=int or number<=0:
        return
    if number%2 == 0:
        return "짝수"
    else:
        return "홀수"
```

```python
print(odd_even(0))
```

```python
def check_leap_year(year):
    if year%4==0 and (not (year%100==0 and year%400!=0)):
        return "윤년"
    return "평년"
```

```python
def day(month):
    month31=[1,3,5,7,8,10,12]
    if month in month31:
        return 31
    elif month==2:
        return 28
    else:
        return 30
```
