## 백준 1931) 회의실 배정

```python
import sys
import heapq
# input = sys.stdin.readline

meets = [] # 가능한 회의 시간
n = int(input())

for i in range(n):
    s, e = map(int, input().split())
    heapq.heappush(meets,(e, s)) # (end time, start time) 전달: end가 작은 순으로 조사해야 더 많은 회의를 담을 수 있음

def solution(meets):
    last_meet = heapq.heappop(meets) # 첫 회의(끝나는 시간이 가장 빠른)
    cnt = 1
    
    while meets:
        current = heapq.heappop(meets) # end가 가장 빠른 회의를 꺼내서
        if current[1] >= last_meet[0]: # 이전 회의의 end보다 start가 빠르면 continue
            last_meet = current
            cnt += 1
    return cnt
print(solution(meets))
```
