```python
n=int(input("2부터 입력하신 수까지의 모든 소수를 찾는 프로그램입니다.\n찾는 범위를 입력하세요 :"))
l=[2,3]

for i in range(2,n+1):
    for j in range(2,int(i**(1/2))+1):
        if i%j==0:
            break
        if j==int(i**(1/2)):
            l.append(i)
i=0
line=0
while i < len(l):
    print(l[i],end=' ')
    line +=1
    if line==7:
        print()
        line=0
    i+=1
```

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}

# set.union(자료형1, 자료형2) 합집합
print(set.union(A,B))
print(A | B)

# set.differnce(자료형1, 자료형2) 차집합
print(set.difference(A,B))
print(A - B)

#set.intersection(자료형1,자료형2) 교집합
print(set.intersection(A,B))
print(A & B)
```

```python
# 자료형1.update(자료형2): 자료형1에 자료형2를 추가(중복 제거)
A.update(B) # A |= B -> A에 A와 B의 합집합 대입
print(A)
print(B)
```

```python
# 자료형1.intersection_update(자료형2): 자료형1에 자료형2와의 교집합 대입
A={1,2,3,4}
B={3,4,5}
A.intersection_update(B) # A &= B -> A에 A와 B의 교집합 대입
print(A)
print(B)
```

```python
# 자료형1.difference_update(자료형2): 자료형1에서 자료형2와의 교집합을 뺀 부분
A= {_ for _ in range(1,5)}
B= {_ for _ in range(2,6) if _ !=3}
print(A,B)
A.difference_update(B) # A -= B
print(A)
```

```python
# 자료형1.symmetric_difference_update(자료형2): 자료형1과 자료형2의 합집합에서 교집합을 뺀 부분
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ !=3}
print(A,B)
A.symmetric_difference_update(B) # A ^= B
print(A)
```

```python
# 자료형1.issubset(자료형2): 자료형1이 자료형2의 부분집합인지 여부(bool)
#     -> A가 B에 포함되는 경우 + A와 B가 일치하는 경우
A = {_ for _ in range(0,3)}
B = {_ for _ in range(0,3)} 
print(A.issubset(B))
```

```python
# 자료형1.issuperset(자료형2): 자료형1이 자료형2를 포함하는 경우(<->issubset)
A= {_ for _ in range(3)}
B= {_ for _ in range(2)}
print(A.issuperset(B))
```

```python
# 자료형1.isdisjoint(자료형2): 교집합이 없으면 True 반환
A= {_ for _ in range(0,3)}
B= {_ for _ in range(3,5)}
A.isdisjoint(B)
```
