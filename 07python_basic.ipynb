```python
while True:
    m,n = map(int,input("정수 두 개 입력하세요:").split(' '))
    if m<1 or m >20:    # not 1 <= m <= 20
        print("첫 번째 값의 범위는 1 ~ 20입니다")
        continue
    elif n<10 or n>30:
        print("두 번째 값의 범위는 10 ~ 30입니다")
        continue
    elif m>n:
        print("첫 번째 입력 값은 두 번째 입력 값보다 작게 입력하세요.")
        continue
    else:
        break
list_a=[2**i for i in range(m,n+1) if i!=m+1 and i!=n-1]
print(list_a)
```

```python
n=int(input("2부터 입력하신 수까지의 모든 소수를 찾는 프로그램입니다.\n찾는 범위를 입력하세요 :"))
l=[2,3]

for i in range(2,n+1):
    for j in range(2,int(i**(1/2))+1):
        if i%j==0:
            break
        if j==int(i**(1/2)):
            l.append(i)
i=0
line=0
while i < len(l):
    print(l[i],end=' ')
    line +=1
    if line==7:
        print()
        line=0
    i+=1
```

```python
n=int(input("2부터 입력하신 수까지의 모든 소수를 찾는 프로그램입니다.\n찾는 범위를 입력하세요 :"))
i = 2
line = 0

while i <= n : 
    j=2
    while j<=i:
        if i% j == 0:
            break
        j+=1
    if j ==i:
        print(i, end=' ')
        line+=1
        if line == 7:
            line = 0
            print()
    i+=1
```

```python
ran = random.randint(1,100)
count=0
record=[]
while True:
    count+=1
    user = int(input())
    record.append(str(user))
    if not 1<= user <=100:
        print("1~100 사이의 숫자를 입력하세요.")
    elif user>ran:
        print("그보다 작습니다.")
    elif user<ran:
        print("그보다 큽니다.")
    else:
        print("맞췄습니다.")
        print(f"{count}번 만에 맞추셨습니다.")
        print("입력한 숫자는 다음과 같습니다.",",".join(record))
        break
    
        
```

```python
A = {_ for _ in range(1,5)}
B = {_ for _ in range(2,6) if _ != 3}

# set.union(자료형1, 자료형2)
print(set.union(A,B))
print(A | B)

# set.differnce(자료형1, 자료형2)
print(set.difference(A,B))
print(A - B)
```

```python
# 자료형1.update(자료형2):
A.update(B) # A |= B
print(A)
print(B)
```

```python
# 자료형1.intersection_update(자료형2)
A.intersection_update(B) # A &= B
print(A)
print(B)
```
