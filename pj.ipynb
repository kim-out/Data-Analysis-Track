```python
def clean(a):
    if ',' in a and '(' in a :
        if a.index(',') >=  a.index('('):
            return a[:a.index('(')]
        else:
            return a[:a.index(',')]
    elif ',' in a:
        return a[:a.index(',')]
    elif '(' in a:
        return a[:a.index('(')]
l = list(map(clean,spatials))
l
```
